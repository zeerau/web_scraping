```python
import requests
from bs4 import BeautifulSoup
from fake_useragent import UserAgent
import concurrent.futures
import pandas as pd
```

```python
url = 'https://www.jumia.com.ng/laptops/'
ua = UserAgent()
userAgent = ua.random
headers = {'User-Agent': userAgent}

page = requests.get(url, headers = headers)
soup = BeautifulSoup(page.content,'html.parser')
```

```python
movie_box = soup.find_all('div', class_ = "info")

Product_name = []
new_price = []
old_price = []
rate_votes = []

for box in movie_box:

    #titles
    if box.find('h3', class_ = "name") is not None:
        title = box.find('h3', class_ = "name")
        Product_name.append(title.text)
    else:
        Product_name.append('None')

    #Numbers
    if box.find('div', class_= "prc") is not None:
        number = box.find('div', class_ = "prc")
        new_price.append(number.text)
    else:
        new_price.append('None')
    
    #Numbers
    if box.find('div', class_= "old") is not None:
        number = box.find('div', class_ = "old")
        old_price.append(number.text)
    else:
        old_price.append('None')

    #rate_votes
    if box.find('div', class_ = "stars _s") is not None:
        rate_vote = box.find('div', class_ = "stars _s")
        rate_votes.append(rate_vote.text)
    else:
        rate_votes.append('None')

movie_df = pd.DataFrame({
    'Product_Name':Product_name,
    'Price': new_price,
    'Old_Price': old_price,
    'No_of_stars':rate_votes
})
movie_df
```
